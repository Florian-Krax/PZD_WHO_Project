```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
values = np.linspace(-1, 1, 1000) ** 2
bins = np.linspace(0,1,20)

plt.hist(values, bins=bins)
plt.show()
```

```python
pvalues = np.array([3,4,7,6,13])
plt.pie(pvalues)
plt.show()
```

```python
values = np.array([4,7,6,3.5,11])

plt.bar(['1', '2', '3', '4', '5'],values)
plt.show()
```
